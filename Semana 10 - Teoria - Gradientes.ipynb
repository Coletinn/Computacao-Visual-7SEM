```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
############### GRADIENTE ROBERTS ###############

import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# Lê a imagem
img = cv2.imread('cachorro.jpg', 0)

# Aplica o operador de gradiente Roberts na horizontal
roberts_x = cv2.filter2D(img, -1, np.array([[-1, 0], [0, 1]]))

# Aplica o operador de gradiente Roberts na vertical
roberts_y = cv2.filter2D(img, -1, np.array([[0, -1], [1, 0]]))

# Calcula a magnitude do gradiente
roberts_mag = np.sqrt(roberts_x ** 2 + roberts_y ** 2)

# Plota as imagens
plt.subplot(2, 2, 1), plt.imshow(img, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 2), plt.imshow(roberts_x, cmap='gray')
plt.title('Gradiente Horizontal'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 3), plt.imshow(roberts_y, cmap='gray')
plt.title('Gradiente Vertical'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 4), plt.imshow(roberts_mag, cmap='gray')
plt.title('Magnitude do Gradiente'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
########## GRADIENTE PREWITT ###########

# leitura da imagem
img = cv2.imread('cachorro.jpg', cv2.IMREAD_COLOR)

# criação dos filtros Prewitt
kernelx = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
kernely = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

# aplicação dos filtros utilizando a função filter2D
img_prewittx = cv2.filter2D(img, -1, kernelx)
img_prewitty = cv2.filter2D(img, -1, kernely)

# cálculo da magnitude do gradiente
img_prewitt = np.sqrt(img_prewittx**2 + img_prewitty**2)

# exibição das imagens
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(img_prewittx, cv2.COLOR_BGR2RGB))
plt.title('Operador Prewitt na horizontal')
plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(cv2.cvtColor(img_prewitty, cv2.COLOR_BGR2RGB))
plt.title('Operador Prewitt na vertical')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
############ GRADIENTE SOBEL #############
```

```python
# aplicar o operador de gradiente Sobel horizontal
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

# aplicar o operador de gradiente Sobel vertical
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# calcular a magnitude do gradiente
mag = np.sqrt(sobelx**2 + sobely**2)

# mostrar as imagens resultantes
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,4),plt.imshow(mag,cmap = 'gray')
plt.title('Magnitude do Gradiente'), plt.xticks([]), plt.yticks([])

plt.show()
```
